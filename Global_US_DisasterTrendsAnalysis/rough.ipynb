```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {'Category': ['A', 'B', 'C', 'D'],
        'Values': [40, 30, 20, 10]}
df = pd.DataFrame(data)

# Custom function to display actual values
def absolute_value(val):
    index = list(df['Values'] * 100.0 / df['Values'].sum()).index(val)  # Find index of val
    return f"{df['Values'][index]}"

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(df['Values'], labels=df['Category'], autopct=lambda p: f'{int(p * sum(df["Values"]) / 100)}', startangle=140)

# Show the pie chart
plt.title("Pie Chart Showing Actual Values")
plt.show()
```
